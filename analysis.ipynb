```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("soil_sensor_data.csv", parse_dates=['Timestamp'])
```

```python
# Create subplots
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(15, 10))

# Plot each variable
axs[0, 0].plot(df['Timestamp'], df['Temperature (C)'], color='r')
axs[0, 0].set_title('Temperature (C)')

axs[0, 1].plot(df['Timestamp'], df['Conductivity (us/cm)'], color='b')
axs[0, 1].set_title('Conductivity (us/cm)')

axs[1, 0].plot(df['Timestamp'], df['Humidity (%)'], color='g')
axs[1, 0].set_title('Humidity (%)')

axs[1, 1].plot(df['Timestamp'], df['Nitrogen (mg/kg)'], color='purple')
axs[1, 1].set_title('Nitrogen (mg/kg)')

axs[2, 0].plot(df['Timestamp'], df['pH'], color='orange')
axs[2, 0].set_title('pH')

axs[2, 1].plot(df['Timestamp'], df['Phosphorus (mg/kg)'], color='cyan')
axs[2, 1].set_title('Phosphorus (mg/kg)')

# Hide empty subplot
fig.delaxes(axs[2, 1])

# Improve layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Timestamp'], df['Temperature (C)'], color='r')
plt.title('Temperature (C)')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
